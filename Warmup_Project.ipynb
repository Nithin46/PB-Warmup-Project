**TASK-2**


count of each category appeared

```python
top5 = {}
for i in text_file.collect():
    if i.startswith("URL"):
        category = i.split('/')[6]
        if category in top5:
            top5[category] += 1
        else:
            top5[category] = 1
```

Top 5 categories

```python
top5_cat = list(dict(Counter(top5).most_common(5)).keys())
```

Making url as key and news in the article as a value

```python
current_url = ""
url = {}
for i in nonempty_lines.collect():
    if i.startswith("URL") :
        current_url = i
    else:
        if current_url in url:
            url[current_url] += i
        else:
            url[current_url] = i
```

Top 5 categories paragraphs

```python
top5_info = ""
for article in url:
    if any(top in article for top in top5_cat):
        top5_info += url[article]
top5_wc = top5_info.split(" ")
top5_wc = [(i.lower()).replace(".", "").replace('”', "").replace(',', "").replace("“", "") for i in top5_wc]
top5_wc_exclude_sw = [i for i in top5_wc if i not in stop_words]
top5_cat_freq = {}
for i in top5_wc_exclude_sw:

    if i not in top5_cat_freq:
        top5_cat_freq[i] = 1
    else:
        top5_cat_freq[i] += 1
```

Top 5 categories wordcloud

```python
top5cattop100 = dict(Counter(top5_cat_freq).most_common(100))
```

Plotting in wordcloud for top5 category words

```python
wordcloud.generate_from_frequencies(frequencies=top5cattop100)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
